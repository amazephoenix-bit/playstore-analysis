###IMPORTING DATA

```python
import pandas as pd
df1=pd.read_csv("/content/googleplaystore.csv")
df2=pd.read_csv("/content/googleplaystore_user_reviews.csv")
```

###CLEANING DATASETS



```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
df1['Rating'].fillna(0,inplace=True)
df1['Current Ver'].fillna('NA',inplace=True)
df1['Android Ver'].fillna('NA',inplace=True)
df1['Content Rating'].fillna('NA',inplace=True)
df1['Type'].fillna('NA',inplace=True)
df1.isnull().sum()

```

```python
df2['Translated_Review'].fillna('NA',inplace=True)
df2['Sentiment'].fillna('Netural',inplace=True)
df2['Sentiment_Polarity'].fillna(0,inplace=True)
df2['Sentiment_Subjectivity'].fillna(0,inplace=True)
df2.isnull().sum()
```

```python
df1 = df1[df1['Installs'] != 'Free']
```

```python
import numpy as np
df1['Installs'] = df1['Installs'].str.replace('[+,]', '', regex=True).astype(int)
df1['Price'] = df1['Price'].str.replace('$', '').astype(float)
df1['Reviews'] = df1['Reviews'].astype(int)
```

###MERGE


```python
df= df1.merge(df2, on='App', how='inner')
df.head()
```

```python
df.shape
```

###ANALYSES

```python
import matplotlib.pyplot as plt

data = df.groupby('Category')['Installs'].mean().sort_values(ascending=False)

plt.figure()
data.plot(kind='bar')

plt.title('Average Installs by Category')
plt.xlabel('Category')
plt.ylabel('Average Installs')

plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('average_installs_by_category.png')
plt.show()
```

```python
plt.figure()
plt.scatter(df['Rating'], df['Installs'])

plt.yscale('log')

plt.xlabel('Rating')
plt.ylabel('Installs (log scale)')
plt.title('Rating vs Installs')
plt.savefig('rating_vs_installs.png')
plt.show()
```

```python
df.groupby('Type')['Installs'].mean().plot(kind='bar')
plt.yscale('log')
plt.title('Average Installs: Free vs Paid')
plt.ylabel('Avg Installs')
plt.savefig('avg_installs_free_vs_paid.png')
plt.show()
```

```python
df.groupby('Type')['Rating'].mean().plot(kind='bar')
plt.title('Average Rating: Free vs Paid')
plt.ylabel('Avg Rating')
plt.savefig('avg_rating_free_vs_paid.png')
plt.show()
```
